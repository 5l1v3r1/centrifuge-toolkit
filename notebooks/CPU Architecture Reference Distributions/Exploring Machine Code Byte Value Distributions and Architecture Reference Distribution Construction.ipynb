<h1>Introduction</h1>

Is the distribution of byte values of machine code in binary executable files targeting the same architecture generally consistent? In other words, is it consistent regardless of outside factors, such as the size or specific purpose of the file?

It should be noted here that the following factors will not be considered due to lack of data pertaining to architectures that are not x86 or AMD64:
 - impact of source language being C versus C++ on machine code byte value distributions
 - influence of compiler toolchain on machine code byte value distributions

In the previous notebooks, it was implied that the distribution of byte values in a given blob of machine code depends primarily on the target CPU family of said machine code. This will be made plain by examining several samples from files targeting 2 different architectures - AMD64 little endian (aka x86-64) and MIPS 32-bit little endian. A more diverse set of architectures and larger number of files will be investigated in subsequent notebooks.

For the sake of simplicity, these files can be renamed:

 2. `d7a281d814021e82cd3b07fb91e5bd94` = "amd64-b"
 3. `d665ddd1ec3e1154898ae5292ae8a963` = "amd64-c"
 
Since these are all ELF binaries with intact section headers, it is possible to locate and extract the `.text` section of each file. The `.text` section contains machine code.

```
Section Headers:
  [Nr] Name              Type            Address          Off    Size   ES Flg Lk Inf Al
...
  [13] .text             PROGBITS        000000000040b550 00b550 130dea 00  AX  0   0 16    # amd64-a
  [13] .text             PROGBITS        00000000000042d0 0042d0 035d42 00  AX  0   0 16    # amd64-b
  [13] .text             PROGBITS        0000000000403160 003160 008a62 00  AX  0   0 16    # amd64-c

```

`pyelftools` can be used to parse the files and read the content of their respective `.text` sections:

````python
import os
import sys

sys.path.append("../../pyelftools")             # relative location of pyelftools repo cloned from github
from elftools.elf.elffile import ELFFile

amd64_code_dict = {}                          # store extracted code together with file name in a dictionary
amd64_path = "../files/ELF/amd64/"

files = os.listdir(amd64_path)
for filename in files:
    with open(amd64_path + filename, "rb") as f:
        elffile = ELFFile(f)
        text_section_header = elffile.get_section_by_name(".text").header
        offset = text_section_header.sh_offset
        size = text_section_header.sh_size
   
        f.seek(offset)
        amd64_code_dict[filename] = list(f.read(size))
````

````python
# quick check

for key,value in amd64_code_dict.items():
    print(key + "\t" + hex(len(value)) + " " + str(value[0:10]))
````

The output above indicates that the machine code has been successfully extracted from each file and converted to a list of integers. We can procede with analysis.

To begin, a density plot of the byte values in the extracted machine code for each file will be shown individually, then overlaid onto the same plot:

````python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 9]
import seaborn as sns
sns.set_style("whitegrid")

# separate plot for each file
for file in sorted(files):
    ax = sns.distplot(amd64_code_dict[file], # list of byte values corresponding to file name
                      bins = 256,
                      hist = False)
    ax.set_title(file + "    machine code size (in bytes): " + str(len(amd64_code_dict[file])))
    ax.set_ylim(0, 0.015)
    plt.show()

# curves overlaid on same plot
for file in sorted(files):
    ax = sns.distplot(amd64_code_dict[file], # list of byte values
                      bins = 256,
                      hist = False)
plt.legend(labels=sorted(files))
#ax.legend(sorted(files))
plt.show()
````

Based on the plots above, it appears that as the amount of machine code increases, the relative frequencies of certain values also increase. The height and sharpness of the peaks in the kernel density estimates decrease as the amount of machine code deacreases.

<h1>2. Constructing an example AMD64 machine code reference distribution</h1>

While the characteristics of the above machine code above are largely consistent, size (number of bytes) does seem to have some affect on the distribution of byte values. When building a reference distribution that captures to a reasonable level of accuracy the similarities and differences in machine code byte distribtutions across files of varying size and purpose, one approach is taking many equal-sized random samples of machine code from a variety of different files. Measuring the distance between a test distribution and this reference distribution provides an objective measure of similarity that can be used to determine whether a code sample is AMD64 or targets some other architecture.

````python
import pandas as pd
import numpy as np
import random
from tqdm import tqdm

random.seed(1)
````

````python
# first, sort byte lists
for file in files:
    amd64_code_dict[file] = sorted(amd64_code_dict[file])
````

To better understand how sampling works, the results of the sampling process can be visualized.

````python
def plot_sample(sample):
    sns.distplot(sample,
                 norm_hist=True,     
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':0.1, 'alpha':0.1},
                 kde_kws={'cumulative': True},  # want CDF
                 bins=256,                      
                 color="blue")
    

def create_sample_df(sample_size, n_rows, sample_source, plot=True):
    
    samples_df = pd.DataFrame(index=np.arange(0, n_rows), columns=[i for i in range(sample_size)])

    for i in np.arange(0, n_rows):
        sample = random.sample(sample_source, sample_size)
        samples_df.loc[i] = sorted(sample) # extremely important: every sample sorted 0 - 255 first, then written to df
        if plot is True:
            plot_sample(sample)
    
    samples_mean = samples_df.mean(axis=0) # axis 0 : rows. Mean of of every column taken
   
    if plot is True:
        sns.distplot(samples_mean, norm_hist=True, kde=False, hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':1, 'alpha':1},kde_kws={'cumulative': True},bins=256,color="red")
        plt.show()
    
    return samples_df, samples_mean
````

In the CDF plots below, the blue lines are CDFs of the random samples taken of the machine code of the specified file. The red line is a CDF of the means of all the random samples. Samples of size 100, 500 and 1000 are taken.

````python
create_sample_df(100, 1000, amd64_code_dict["amd64-a"])
create_sample_df(500, 1000, amd64_code_dict["amd64-a"])
amd64_a_sample_df, amd64_a_sample_mean = create_sample_df(1000, 1000, amd64_code_dict["amd64-a"])
````

As sample size increases, the average of all samples (the red line on the plots above) gets closer and closer to the CDF created from all the bytes in the `.text` section of `amd64-a`. In other words, by taking many small samples (comparatively speaking) of the byte values in the machine code and averaging the results together, the overall distribtution of byte values in the machine code can be estimated quite accurately. We can see just how accurate this 1000-byte estimate is by comparing it with the full 1248746-byte distribution.

````python
# plot CDF of all bytes values in .text section of amd64-a
ax = sns.distplot(amd64_code_dict["amd64-a"], 
                  bins = 256,
                  kde=False,
                  hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':1, 'alpha':1}, 
                  kde_kws={'cumulative': True},
                  norm_hist=True,
                  color="blue")

# next, plot CDF of sample means
ax = sns.distplot(amd64_a_sample_mean, 
                  bins = 256,
                  kde=False,
                  hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':1, 'alpha':1}, 
                  kde_kws={'cumulative': True},
                  norm_hist=True,
                  color="red")
ax.set_title(file)
plt.legend(["CDF of full distribution", "CDF of mean of samples"], loc="upper left")
plt.show()
````

As we can see, the 2 CDFs above are practically on top of each other. The degree to which these two distributions differ can be objectively quantified by measuring the Wasserstein distance between them.

````python
from scipy import stats

stats.wasserstein_distance(amd64_code_dict["amd64-a"], amd64_a_sample_mean)
````

Another metric that can be used is energy distance:

````python
stats.energy_distance(amd64_code_dict["amd64-a"], amd64_a_sample_mean)
````

Since this is the first time we have measured the Wasserstein distance between 2 distributions, we have no frame of reference to compare this result to. However, we can guess that this is a small distance based on how close the CDFs are in the plot above.

The next step is building a reference distribution representative AMD64 machine code against which test distributions can be measured. To do this, the mean of samples taken from `amd64-a`, `amd64-b` and `amd64-c` will be averaged together.

````python
amd64_b_sample_df, amd64_b_sample_mean = create_sample_df(1000, 1000, amd64_code_dict["amd64-b"], plot=False)
amd64_c_sample_df, amd64_c_sample_mean = create_sample_df(1000, 1000, amd64_code_dict["amd64-c"], plot=False)
````

````python
amd64_sample_means_df = pd.DataFrame([amd64_a_sample_mean, 
                                      amd64_b_sample_mean,
                                      amd64_c_sample_mean],
                                      columns=[i for i in range(1000)])
amd64_reference_distribution = amd64_sample_means_df.mean(axis=0)
````

Let us compare the means of samples from the 3 files as well as the newly-created example reference distribution.

````python
amd64_dists = [amd64_a_sample_mean, amd64_b_sample_mean, amd64_c_sample_mean]
for dist in amd64_dists:
    sns.distplot(dist, 
                 bins = 256,
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':1.5, 'alpha':0.5}, 
                 kde_kws={'cumulative': True},
                 norm_hist=True)
    
sns.distplot(amd64_reference_distribution, 
             bins = 256,
             kde=False,
             hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':1, 'alpha':1}, 
             kde_kws={'cumulative': True},
             norm_hist=True,
             color="red")

plt.legend(["amd64-a samples mean", "amd64-b samples mean", "amd64-c samples mean", "amd64 reference distribution"], loc="upper left")
plt.show()
````

Now that a reference distribution exists for AMD64 machine code, any distribution can be compared with it in order to determine that distribution's statistical similarity to AMD64 machine code. It makes sense to begin with the distributions representative of the machine code from `amd64-a`, `amd64-b` and `amd64-c`. The CDFs in the plot above indicate the Wasserstein distance between their representative distributions and the AMD64 reference distribution should be quite small.

````python
for dist in amd64_dists:
    print(stats.wasserstein_distance(amd64_reference_distribution, dist))
````

Average: `3.358376889`

The above Wasserstein distances are not quite as small as the Wasserstein distance between the machine code byte value distribution of `amd64-a` and its representative sample, but still quite small.

Below are the energy distances between the AMD64 reference distribution and the `amd64-a`, `amd64-b`, and  `amd64-c` sample distributions. These distances are fairly similar to the Wasserstein distances calculated above in terms of information conveyed about the distances between the distributions.

````python
for dist in amd64_dists:
    print(stats.energy_distance(amd64_reference_distribution, dist))
````

Average: `0.352975198`

The more 2 distributions deviate from each other, the farther the resulting line will be from the perfect 0,0 - 255,255 diagonal.

````python
ax = plt.plot(amd64_reference_distribution, amd64_reference_distribution)
plt.title("Perfect match")
plt.show()
````
