<h1>Introduction</h1>

The Centrifuge toolkit leverages existing Python libraries to provide facilities for visualizing information in binary files. The approach to analysis taken by this toolkit is based on file partitioning: when a binary file is split into blocks, statistical measurements of each block are taken; these per-block measurements then can be plotted to explore the different types of information that may be present in a binary. 

Although development of this toolkit was prompted by a desire to analyze executable binaries (firmware in particular) with statistical methods and machine learning techniques more thoroughly than was possible with previously existing tools, Centrifuge is format-agnostic - it treats the bytes in a file as raw integer values without any interpretation and does not rely on any kind of metadata or information about how a file may be structured. As a result, it can be used to analyze any type of information encoded in 8-bit bytes, not just files containing machine code.

In this notebook, a demonstration will be given of the various ways in which information within a binary file can be visualized with Centrifuge's plotting functions. The subject of the analysis will be the Linux coreutils program `bash`. These visualizations will also serve to provide support for the hypothesis underlying Centrifuge's creation, namely that it is possible to automatically identify and extract machine code from an arbitrary binary executable using *unsupervised* machine learning.

<h3>Brief Overview</h3>

Information encoded in a binary file can be analysed in a variety of ways:

 - treating the entire file as a blob of bytes [1]
 - relying on metadata such as a file header or section header table to extract different types of information directly [2]
 - using a "sliding window" [3][4][5][6]
 - partitioning the file into same-sized pieces or blocks [7]
 
The last approach, partitioning a file into blocks, is the one implemented by Centrifuge. This was inspired by the method used by [binwalk](https://github.com/ReFirmLabs/binwalk) to [measure file entropy](https://github.com/ReFirmLabs/binwalk/blob/ff34b1219e9b34bcd47251f63ba99e12d01bfc74/src/binwalk/modules/entropy.py#L150). However, aside from entropy, Centrifuge measures an additional 6 features per block. None of the analyses performed by this toolkit would have been possible without these engineered features. This approach makes it straightforward to use statistical methods or machine learning algorithms to analyze the information encoded in a binary file.

![overview](approach.png)

Mean, median, standard deviation, and entropy are self-explanatory. 
 - `Standard deviation of counts` is the standard deviation of the counts (frequency) of the byte values in each block. The distribution of byte value counts is different from the distribution of byte values themselves and as a result provides additional information. 
 - `ASCII ratio` is the proportion of bytes in the printable ASCII range (0x20 - 0x7e). Use of this variable was prompted by [cortesi's](https://github.com/cortesi) [binvis.io](http://binvis.io/#/) as well as exploratory data analysis of binary files via eaxmination of hex dumps, in which it was observed that different types of information (e.g. machine code vs. hardcoded strings) had distinctly different proportions of printable ASCII bytes.
 - `0x00 ratio` is simply the proportion of bytes in a block that have a value of zero. This also varies based on the information encoding; blocks of compressed or encrypted data or text data have few 0x00 bytes, whereas blocks containing machine code will have more.
 
These 7 features provide a way to quantify and objectively measure properties of the information stored in a binary file. Through visualization of these measurements, it is possible to gain insights into how information in binary files is organized that would be more difficult to attain using other means of analysis. An example of one such insight is that it is possible to use unsupervised machine learning techniques (specifically, density-based clustering) to distinguish between different types of information e.g. machine code versus non-machine code information. To demonstrate why this might be the case, we can use Centrifuge to visualize the information in executable binaries like `bash` and observe the patterns that emerge.

It should be noted here that it is certainly possible to engineer even more features to use for analysis. This is an area suitable for future experimentation.

<h1>Basic Visualization</h1>

````python
import sys
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [16, 9]  # we want large, uniformly-sized plots

sys.path[0:0] = ['.', '..']               # the centrifuge library files are located in ../centrifuge

from centrifuge.binfile import BinFile
````

To create plots of information in a binary, a `BinFile` object must first be instantiated from the target file. Once this is done, the various methods available can be viewed. Four in particular have to do with plotting:
 - `plot_file_entropy`
  - similar to the entropy plotting functionality offered by `binwalk`, but more flexible
 - `plot_variables_by_range`
  - plot any 2 of the engineered variables against each other. The shapes of the plot markers, their colors, and the plot labels can be specified
 - `show_scatter_matrix`
  - this creates a 7x7 matrix of scatterplots
 - `plot_DBSCAN_results`
  - If clusters are found by DBSCAN, they are overlayed onto an entropy plot of the file to provide context for the analyst

````python
file_handle = open("/bin/bash", "rb")
binfile = BinFile(file_handle)

dir(binfile)[-28:]
````

Before any analysis or visualization can commence, the file must first be partitioned into blocks. This is always the first step. The size of the blocks can be specified by the user. A block size of 1024 was used in all experiments and will be used here.

````python
binfile.set_block_size(1024)
binfile.slice_file()
````

Now that the file has been partitioned into blocks and each feature has been measured, a data frame containing all of this information can be examined:

````python
binfile.file_data_frame
````

As we can see, `bash` has been partitioned into 1088 blocks, and for each block, seven different features have been measured. Now that the information in the binary file has been quantified, it can be explored through visualization and machine learning methods can be used to analyze it in order to find interesting patterns.

<h2>Plotting Any Single Feature</h2>

Entropy plots are features of many tools, not just `binwalk`. However, there are many features besides entropy that can give us insight into file information organization. 

Any of the features measured for each block can be plotted. Below are plots of 
 - block means
 - medians
 - byte value standard deviations
 - byte value counts standard deviations
 - printable ASCII ratios
 - 0x00 byte ratios
 
The `.text` section is demarcated throughout:

````python
plot_features = {"mean": "royalblue",
                 "median": "darkgoldenrod",
                 "std_dev": "forestgreen",
                 "std_dev_counts": "lightseagreen",
                 "ascii": "purple",
                 "zeroes": "black"}

for feature, color in plot_features.items():
    binfile.plot_file_feature(feature, color, 0x02cbc0, 0x02cbc0+0x0a2c02)
````



<hr>





<h2>2D Scatterplots</h2>

Naturally, the next step is plotting 2 features against each other, creating a scatterplot in 2-dimensional space. This can be done using `matplotlib` directly. 

The points on the plot represent blocks. In other words, since `bash` was sliced into 1088 blocks, there are 1088 points on each plot below.

````python
plt.scatter(binfile.block_entropy_levels, 
            binfile.block_byteval_std_dev)
plt.title("bash")
plt.xlabel("Block Entropy")
plt.ylabel("Block Standard Deviation")
plt.show()
````

````python
plt.scatter(binfile.block_zeroes_ratios,
            binfile.block_byteval_mean)
plt.title("bash")
plt.xlabel("Block 0x00 ratio")
plt.ylabel("Block Mean")
plt.show()
````

Though clear clusters emerge, it is not immediately obvious what types of information are clustering together. Using Centrifuge, these plots can be recreated such that the points on the plot are shaped and colored based on their location in the file. For example, the blocks that fall within the `.text` section and therefore contain machine code can be represented as red squares, and all other blocks in the file can be represented as black circles:

````python
binfile.plot_variables_by_range(binfile.block_byteval_std_dev,   # 1st variable
                                binfile.block_entropy_levels,  # 2nd variable
                                0x02cbc0,                       # start file offset
                                0x02cbc0+0x0a2c02,              # end file offset
                                target_data_marker='s',         # marker shape for points that fall within the specified offsets
                                other_data_marker='o',          # marker for points outside of offsets
                                target_data_color='red',        # color of points inside offsets
                                other_data_color='black',       # color of points outside offsets
                                title='bash',   
                                xlabel='Block Standard Deviation',
                                ylabel='Block Entropy')
````

````python
binfile.plot_variables_by_range(binfile.block_zeroes_ratios,
                                binfile.block_byteval_mean,
                                0x02cbc0,
                                0x02cbc0+0x0a2c02,
                                target_data_marker='s',
                                other_data_marker='o',
                                target_data_color='red',
                                other_data_color='black',
                                title='bash',
                                xlabel='Block Zeroes Ratio',
                                ylabel='Block Mean')
````

Not only do the blocks containing object code cluster together, but other distinct clusters emerge as well. It seems that clusters form based on information similarity, and that different types of information are organized quite differently. Further investigation provides further evidence of this (points representing blocks from within the `.text` section, which contain machine code, are marked as red squares in all plots):

````python
binfile.plot_variables_by_range(binfile.block_byteval_mean, binfile.block_byteval_median, 0x02cbc0, 0x02cbc0+0x0a2c02, target_data_marker='s', other_data_marker='o', target_data_color='red', other_data_color='black', title='bash', xlabel='Block Mean', ylabel='Block Median')
binfile.plot_variables_by_range(binfile.block_ascii_ratios, binfile.block_byteval_median, 0x02cbc0, 0x02cbc0+0x0a2c02, target_data_marker='s', other_data_marker='o', target_data_color='red', other_data_color='black', title='bash', xlabel='Block ASCII Ratios', ylabel='Block Median')
binfile.plot_variables_by_range(binfile.block_byteval_std_dev, binfile.block_byteval_std_dev_counts, 0x02cbc0, 0x02cbc0+0x0a2c02, target_data_marker='s', other_data_marker='o', target_data_color='red', other_data_color='black', title='bash', xlabel='Block Standard Deviation', ylabel='Block Counts Standard Deviation')
binfile.plot_variables_by_range(binfile.block_zeroes_ratios, binfile.block_ascii_ratios, 0x02cbc0, 0x02cbc0+0x0a2c02, target_data_marker='s', other_data_marker='o', target_data_color='red', other_data_color='black', title='bash', xlabel='Block 0x00 Ratio', ylabel='Block Printable ASCII Ratio')
binfile.plot_variables_by_range(binfile.block_entropy_levels, binfile.block_ascii_ratios, 0x02cbc0, 0x02cbc0+0x0a2c02, target_data_marker='s', other_data_marker='o', target_data_color='red', other_data_color='black', title='bash', xlabel='Block Entropy', ylabel='Block Printable ASCII Ratio')
````

Some clusters are more cohesive than others, but clear patterns emerge. In this file, blocks containing machine code clearly have a tendancy to cluster together.

<h2>Information Visualization via Clustering with DBSCAN</h2>

As we have seen in the plots above, the clusters in the data created from partitioning a file into equal-sized blocks and measuring 7 features of each block vary in shape, size, density and number depending on the information encoded therein. Unsupervised learning algorithms such as DBSCAN can use the measurements of the engineered features to automatically detect information that clusters together (e.g. code or compressed data):

>The DBSCAN algorithm views clusters as areas of high density separated by areas of low density. Due to this rather generic view, clusters found by DBSCAN can be any shape, as opposed to k-means which assumes that clusters are convex shaped. The central component to the DBSCAN is the concept of core samples, which are samples that are in areas of high density. A cluster is therefore a set of core samples, each close to each other (measured by some distance measure) and a set of non-core samples that are close to a core sample (but are not themselves core samples). There are two parameters to the algorithm, `min_samples` and `eps`, which define formally what we mean when we say dense. Higher `min_samples` or lower `eps` indicate higher density necessary to form a cluster. [8]

>**eps: float, default=0.5**
>
>The maximum distance between two samples for one to be considered as in the neighborhood of the other. This is not a maximum bound on the distances of points within a cluster. This is the most important DBSCAN parameter to choose appropriately for your data set and distance function. [9]


>**min_samples: int, default=5**
>
>The number of samples (or total weight) in a neighborhood for a point to be considered as a core point. This includes the point itself. [9]


````python
binfile.cluster_DBSCAN(0.3,                           # eps
                       10,                            # min_samples
                       find_optimal_epsilon=False)    # don't show "knee" plot here
binfile.plot_DBSCAN_results()
````

Looking at the first graph in which markers representing a cluster label are layered on top of an entropy plot of `bash`, we see that the cluster consisting of green upside-down triangles falls neatly within the part of the binary containing the `.text` section, so this is the cluster that represents machine code.

<h2>Analyzing the Clusters</h2>

After each block is assigned a cluster label, it is possible to view a pairwise plot of each feature like in the scatterplot matrix above, but this time with each point colored based on cluster label. This is much clearer and shows off how well DBSCAN performs with the data extracted from the binary based on the engineered features.

````python
# Any block not found to be in any cluster is assigned a cluster label of -1. Here we are not interested 
# in those blocks, and therefore every row that has a value of -1 in the "cluster labels" column is filtered
# out.

# in addition to this, the "core samples mask" column is dropped from the data frame, because this is not a
# feature in the file data.

sns.pairplot(binfile.dbscan_data_frame[binfile.dbscan_data_frame["cluster labels"] != -1].drop(["core samples mask"], axis=1), hue="cluster labels")
````

Each cluster found by DBSCAN can be analyzed individually. The dataframe storing the results of clustering the file data with DBSCAN can be filtered by cluster label. For `eps` = 0.3 and `min_samples` = 10, DBSCAN finds 6 clusters (labelled 0 - 5). A cluster label of -1 means that block was not part of any cluster.

````python
set(binfile.dbscan_data_frame["cluster labels"])
````

````python
binfile.dbscan_data_frame
````

````python
for i in range(0,6):
    print("Cluster %d shape: %s" % (i, binfile.dbscan_data_frame.loc[binfile.dbscan_data_frame['cluster labels'] == i].shape))
````

<h3>Analysis of Each Feature by Cluster</h3>

Cluster 3 is the largest by far, indicating this is the cluster represented by the upside-down green triangles that fall within the `.text` section and therefore contains machine code. Cluster 3 stands out from the rest due to the how variable some of the feature observations are. The ranges of block byte value means, medians, entropy and printable ASCII ratios in cluster 3 are greater than those in the other clusters, sometimes greatly so. 

As was indicated in the DBSCAN results plots above (especially the second plot - Block entropy vs. standard deviation), clusters 0 and 2 behave similarly, as do clusters 1 and 4.

````python
sns.set(style="whitegrid")
sns.set_context("paper")

column_names = list(binfile.dbscan_data_frame)[:-2] # <- here, ignore 'core samples mask', 'cluster labels'

features = {"entropy":8, 
            "zeroes ratios":1, 
            "ascii ratios":1, 
            "byte value std dev":120, 
            "byte value counts std dev":50, 
            "byte value mean":160, 
            "byte value median":160}

for column in column_names:
    ax = sns.violinplot(x="cluster labels", y = column, palette="deep",
                       data=binfile.dbscan_data_frame[binfile.dbscan_data_frame["cluster labels"] != -1]) # exclude rows with cluster label -1
    ymax = features[column]
    ax.set_ylim(-0.025 * ymax, ymax)
    plt.show()
````

<h3>Comparing Distributions of Byte Values by Cluster</h3>

In the violin plots above, each we observed the differences in distributions of features across clusters. It is also possible to analyze the distributions of the values of the bytes in each cluster (each block consists of 1024 bytes sliced from the file. Each cluster consists of many such blocks). Cluster 3 consists of blocks containing machine code and so is of particular interest.

````python
dbscan_df = binfile.dbscan_data_frame
dbscan_df['index_col'] = binfile.dbscan_data_frame.index    # add an index column

colors = ["blue", "orange", "green", "red", "purple", "brown"]
for cluster_label in sorted(list(set(binfile.dbscan_data_frame["cluster labels"])))[1:]:  # [0, 1, 2, 3, 4, 5]
    cluster_indices = dbscan_df[dbscan_df['cluster labels']==cluster_label]['index_col']  # select rows by cluster label
    cluster_blocks = [binfile.blocks[i] for i in cluster_indices]                         # select blocks in list by index
    cluster_data = []
    for block in cluster_blocks:     # for each block, select the 1024 raw bytes
        cluster_data += block.data   # append these bytes to the list
        
    sns.distplot(list(cluster_data), # calling list() on the binary data automatically converts bytes to int values
                 norm_hist=True,     # normalize the y-axis. Different clusters have different numbers of blocks.
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':2, 'alpha':1},
                 kde_kws={'cumulative': True},  # want CDF
                 bins=256,                      # 1 bin per byte value. Maximum granularity
                 color=colors[cluster_label]).set_title("Cluster " + str(cluster_label) + " Cumulative Distribution Function")  # select color based on cluster label
    plt.show()
````

````python
# Same as above, but omitting plt.show() from the loop body causes plots to be layered on top of each other.
# This allows direct visual comparisons of distributions to be made more easily

for cluster_label in sorted(list(set(binfile.dbscan_data_frame["cluster labels"])))[1:]:
    cluster_indices = dbscan_df[dbscan_df['cluster labels']==cluster_label]['index_col']
    cluster_blocks = [binfile.blocks[i] for i in cluster_indices]
    cluster_data = []
    for block in cluster_blocks:
        cluster_data += block.data
        
    sns.distplot(list(cluster_data),
                 norm_hist=True,
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':2, 'alpha':1},
                 kde_kws={'cumulative': True},  
                 bins=256)

plt.legend(["Cluster 0", "Cluster 1","Cluster 2","Cluster 3","Cluster 4","Cluster 5"],
           loc="upper left")
plt.show()
````

The cumulative distribution function plots above clearly show that the clusters found by DBSCAN have unique byte value distributions, including the cluster containing machine code.
