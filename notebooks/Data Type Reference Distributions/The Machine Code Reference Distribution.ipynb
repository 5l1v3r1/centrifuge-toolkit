```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [16, 9]
sns.set_style("whitegrid")

from scipy import stats
```

```python
path = "../centrifuge/distributions/cpu_architectures/"
archs = os.listdir(path)

reference_dict = {}

for file in archs:
    with open(path + file, "rb") as f:
        reference_dict[file] = pickle.load(f)
```

```python
reference_df = pd.DataFrame(index = reference_dict.keys(),
                            columns = [i for i in range(1000)])

for file, code in reference_dict.items():
    reference_df.loc[file] = code
```

```python
for code in reference_dict.values():
    sns.distplot(code,
                 norm_hist=True,     
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':2, 'alpha':1},
                 kde_kws={'cumulative': True},  
                 bins=256)

sns.distplot(reference_df.mean(axis = 0), 
             norm_hist=True,  
             kde=False, 
             hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':10, 'alpha':0.5},
             kde_kws={'cumulative': True},  
             bins=256,
             color="black")

sns.distplot(reference_df.median(axis = 0), 
             norm_hist=True,  
             kde=False, 
             hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':3, 'alpha':0.8},
             kde_kws={'cumulative': True},  
             bins=256,
             color="red")

plt.legend(list(reference_dict.keys())+["mean"]+["median"], loc="upper left")
plt.title("Mean of reference distributions vs. All reference distributions")
plt.show()
```

```python
distance_df = pd.DataFrame(index = reference_dict.keys(),
                           columns = ["Wasserstein Distance (Mean)",
                                      "Wasserstein Distance (Median)"])

for file, code in reference_dict.items():
    distance_df.loc[file]["Wasserstein Distance (Mean)"] = stats.wasserstein_distance(reference_df.mean(axis = 0), code)
    distance_df.loc[file]["Wasserstein Distance (Median)"] = stats.wasserstein_distance(reference_df.median(axis = 0), code)

distance_df.sort_values("Wasserstein Distance (Mean)")
```

```python
with open("../distributions/data_types/machine_code", "wb") as f:
    pickle.dump(reference_df.mean(axis=0), f)
```

Though using the medians minimizes distance overall, using the means might be a better choice. The more distant distributions (MIPSEL, MIPS64EL, ARMEL, PowerPC) are less distant to the means.

```python
def plot_dist_cdf(df):
    for entry in list(df.index):
        sns.distplot(df.loc[entry],
                     norm_hist=True,     
                     kde=False,
                     hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':2, 'alpha':1},
                     kde_kws={'cumulative': True},  
                     bins=256)
    plt.legend(list(df.index), loc="upper left")
    plt.show()
```

```python
def create_family_ref(mc_ref_dist, dist_1, dist_2):
    family_df = pd.DataFrame(index=[dist_1, dist_2],
                             columns=[i for i in range(0,1000)])
    family_df.loc[dist_1] = mc_ref_dist.loc[dist_1]
    family_df.loc[dist_2] = mc_ref_dist.loc[dist_2]
    family_df_median = family_df.median(axis = 0)
    family_df = family_df.append(pd.Series(family_df_median, name="medians"))
    
    return family_df, family_df_median
```
