```python
import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from scipy import stats

plt.rcParams['figure.figsize'] = [16, 9]
```

```python
pool = [i for i in range(0,256)]

samples_df = pd.DataFrame(index=[i for i in range(0, 1000)],
                          columns=[i for i in range(0,1000)])

for i in tqdm(range(0,1000)):
    sample = sorted(np.random.choice(pool, size=1000, replace=True))
    samples_df.iloc[i] = sample
    
means = samples_df.mean(axis = 0)
medians = samples_df.median(axis = 0)
```

```python
def plot_dist_cdf(data):
    sns.distplot(data,
                 norm_hist=True,     
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':2, 'alpha':1},
                 kde_kws={'cumulative': True},  
                 bins=256,
                 color="red")
    plt.show()
    
def plot_dist_hist(data):
    sns.distplot(data,
                 kde=False,
                 hist_kws={'alpha':1},
                 bins=256)
    plt.show()
```

```python
for i in range(0,1000):
    sns.distplot(samples_df.iloc[i],
                 norm_hist=True,     
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':0.3, 'alpha':0.1},
                 kde_kws={'cumulative': True},  
                 bins=256,
                 color="blue")
    
plot_dist_cdf(round(means))
```

```python
plot_dist_hist(round(means))
```

```python
def to_byte_dict(data):
    byte_dict = {}
    for i in range(0, 256): byte_dict.update({i:0})
    for i in list(data): byte_dict[i]+= 1
    return byte_dict
```

```python
d = to_byte_dict(round(means))
stats.entropy(list(d.values()), base=2)
```

```python
d = to_byte_dict(round(medians))
stats.entropy(list(d.values()), base=2)
```

```python
with open("high_entropy", "wb") as f:
    pickle.dump(round(means), f)
```
