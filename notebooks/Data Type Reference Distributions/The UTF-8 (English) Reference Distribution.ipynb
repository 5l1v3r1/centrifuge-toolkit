```python
import os
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
from scipy import stats

plt.rcParams['figure.figsize'] = [16, 9]
sns.set_style("whitegrid")
```

```python
path = "../files/utf8-english/"
files = os.listdir(path)
files.remove('README.txt')
files
```

```python
text_data = {}

for file in files:
    with open(path + file, "rb") as f:
        f.seek(10000)
        text_data[file] = list(f.read(50000))
```

```python
file_means = {}
file_medians = {}

for file, text in text_data.items():
    df = pd.DataFrame(index=[i for i in range(0,1000)],
                      columns=[i for i in range(0,1000)])
    for i in tqdm(range(0, 1000)):
        df.iloc[i] = sorted(np.random.choice(text, size=1000, replace=True))
        
    file_means[file] = df.mean(axis = 0)
    file_medians[file] = df.median(axis = 0)
```

```python
for file, sample in file_means.items():
    sns.distplot(round(sample),
                 norm_hist=True,     
                 kde=False,
                 hist_kws={'histtype':'step', 'cumulative': True, 'linewidth':2, 'alpha':1},
                 kde_kws={'cumulative': True},  
                 bins=256)
    
plt.legend(file_means.keys(), loc="upper left")
plt.show()
```

```python
def plot_dist_hist(data):
    sns.distplot(data,
                 kde=False,
                 hist_kws={'alpha':1},
                 bins=256)
    plt.show()
```

```python
utf8_reference = round(file_means['62875-0.txt'])
plot_dist_hist(utf8_reference)
```

```python
# replace " " with 0

for index, value in enumerate(utf8_reference):
    if value == 32:
        utf8_reference[index] = 0
```

```python
plot_dist_hist(utf8_reference)
```
